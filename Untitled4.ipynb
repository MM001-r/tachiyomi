```python

import gradio as gr
import whisper
import torch
import os
import shutil
import time
import ffmpeg
from transformers import pipeline
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
whisper_model = whisper.load_model("small", device=device)

# Для перевода
translator_en = pipeline("translation", model="Helsinki-NLP/opus-mt-en-ru", device=0 if torch.cuda.is_available() else -1)
translator_ja = pipeline("translation", model="Helsinki-NLP/opus-mt-ja-ru", device=0 if torch.cuda.is_available() else -1)
```

```python
# Форматирование времени для субтитров
def format_timestamp(seconds):
    millisec = int((seconds % 1) * 1000)
    time_struct = time.gmtime(seconds)
    return time.strftime(f"%H:%M:%S,{millisec:03d}", time_struct)

# Автоматический перевод
def translate_text(text, language):
    chunk_size = 500
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    translated = []
    translator = translator_ja if language == 'Japanese' else translator_en
    for chunk in chunks:
        result = translator(chunk)[0]['translation_text']
        translated.append(result)
    return "\n".join(translated)
```

```python
def continue_processing(input_path, transcription, translated_text, segments, nsfw_mode, output_format):
    srt_path = "output_subs.srt"

    # Создаём субтитры
    translated_lines = translated_text.strip().split('\n')
    with open(srt_path, "w", encoding="utf-8") as f:
        for idx, seg in enumerate(segments):
            if idx >= len(translated_lines):
                break
            f.write(f"{idx+1}\n")
            f.write(f"{format_timestamp(seg['start'])} --> {format_timestamp(seg['end'])}\n")
            f.write(f"{translated_lines[idx]}\n\n")

    # Обработка NSFW аудио (затемнение/приглушение)
    if nsfw_mode:
        filtered_audio = "filtered_audio.mp3"
        os.system(f'ffmpeg -y -i "{input_path}" -af "highpass=f=200, lowpass=f=3000" "{filtered_audio}"')
        audio_input = filtered_audio
    else:
        audio_input = input_path

    # Создаём озвучку (пока простым способом)
    output_audio = "dubbed_audio.wav"
    os.system(f'ffmpeg -y -i "{audio_input}" -vn -acodec pcm_s16le -ar 44100 -ac 2 "{output_audio}"')

    final_output = "final_output" + output_format

    if output_format == ".mp4":
        # Встраиваем субтитры и новую озвучку
        os.system(f'ffmpeg -y -i "{input_path}" -i "{output_audio}" -vf subtitles="{srt_path}" -map 0:v -map 1:a -c:v libx264 -c:a aac "{final_output}"')
    else:
        shutil.copy(output_audio, final_output)

    return None, None, final_output, srt_path, None
```

```python
with gr.Blocks() as app:
    gr.Markdown("## Перевод, дубляж и субтитры видео/аудио файлов на русский язык")

    with gr.Row():
        file_input = gr.File(label="Загрузите файл (видео или mp3)")
        language_input = gr.Dropdown(["Auto", "English", "Japanese"], value="Auto", label="Язык оригинала")
        nsfw_checkbox = gr.Checkbox(label="Режим NSFW/АСМР?")
        output_format_input = gr.Dropdown([".mp4", ".mp3"], value=".mp4", label="Формат результата")
        manual_edit_checkbox = gr.Checkbox(label="Редактировать перевод вручную?")

    submit_btn = gr.Button("Запустить первичную обработку")

    with gr.Row():
        transcription_box = gr.Textbox(label="Транскрипция оригинала", visible=False, lines=10)
        translation_box = gr.Textbox(label="Перевод на русский", visible=False, lines=10)

    continue_btn = gr.Button("Продолжить обработку после редактирования", visible=False)

    final_output_file = gr.File(label="Финальный файл", visible=False)
    srt_output_file = gr.File(label="Файл субтитров", visible=False)

    # Скрытые переменные
    segments_state = gr.State()

    # Основная обработка
    def wrapped_process_upload(file, language, nsfw_mode, output_format, manual_edit):
        transcription, translation, final_file, srt_file, segments = process_upload(file, language, nsfw_mode, output_format, manual_edit)
        return transcription, translation, final_file, srt_file, segments, gr.update(visible=manual_edit)

    submit_btn.click(
        wrapped_process_upload,
        inputs=[file_input, language_input, nsfw_checkbox, output_format_input, manual_edit_checkbox],
        outputs=[transcription_box, translation_box, final_output_file, srt_output_file, segments_state, continue_btn]
    )

    # Продолжение после редактирования
    continue_btn.click(
        continue_processing,
        inputs=[file_input, transcription_box, translation_box, segments_state, nsfw_checkbox, output_format_input],
        outputs=[transcription_box, translation_box, final_output_file, srt_output_file, continue_btn]
    )

app.launch()
```

```python
def process_upload(file, language, nsfw_mode, output_format, manual_edit):
    input_path = file.name
    lang_code = {'Auto': None, 'English': 'en', 'Japanese': 'ja'}

    # Распознавание
    result = whisper_model.transcribe(input_path, language=lang_code.get(language))
    transcription = result['text']
    segments = result['segments']

    # Перевод
    translated_text = translate_text(transcription, language)

    if manual_edit:
        return transcription, translated_text, None, None, None

    return continue_processing(input_path, transcription, translated_text, segments, nsfw_mode, output_format)
```
